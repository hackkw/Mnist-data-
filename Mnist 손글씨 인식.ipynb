```python
import tensorflow as tf #텐서플로우 
import numpy as np #넘파이 라이브러리를 불러옴

import matplotlib.pyplot as plt
%matplotlib inline

mnist = tf.keras.datasets.mnist #케라스에서 제공하는 mnist 데이터 가져오기
(x_train, y_train), (x_test, y_test) = mnist.load_data() # 학습데이터, 평기데이터에 로드

print(f'학습 데이터 {len(x_train)}개 평가 데이터 {len(x_test)}개')

import matplotlib.pyplot as plt #데이터 시각화

row = 10
col = 10
n = row*col

plt.figure(figsize=(5,5)) #그래프 사이즈 설정
for i in range(n):
  ax = plt.subplot(row, col, i+1) #subplot 여러개의 그래프 사용
  plt.imshow(x_train[i].reshape(28,28)) 
  plt.gray() #그레이 변환
  #ax.get_xaxis().set_visible(False)
  #ax.get_yaxis().set_visible(False)
plt.show()

x_train, x_test = x_train / 255.0, x_test /255.0 #0~255 픽셀값으로 이루어진 이미지에 255나누기

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)), #다차원 배열을 1차원 배열로 변환
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10,activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)

scores = model.evaluate(x_test, y_test, verbose=2)
print("정확도도 : %.2f %%"% (scores[1]*100))
pred = model.predict(x_test)
np.set_printoptions(precision=7)
print(f'각 클래스에 속할 확률 : \n{pred[0]}')

from PIL import Image
img_png = Image.open("5.png").convert('L')
img_png.size

img_mnist = img_png.resize((28,28))
img_mnist

np_im = (255 - np.array(img_mnist))
np_im.shape

tt = np.array2string(np_im,separator=",")
tt = tt.replace("\n",".")
tt = tt.replace("[","")
tt = tt.replace("]","")
tt = tt.replace("  ","")
tt = tt.replace(" ","")
tt = "*," + tt
print(tt)

data_value = tt.split(",")
#data_value
image_arr = np.asfarray(data_value[1:]).reshape((28,28))

plt.imshow(image_arr, cmap='Greys', interpolation='None')
plt.show()

image_array = np.asfarray(data_value[1:]).reshape(1,28,28)

predictions = model(image_array).numpy()
result = np.argmax(tf.nn.softmax(predictions).numpy())
print(f'나의 이미지 분류 결과: {result}')




```
